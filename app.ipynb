```python
from flask import Flask, render_template, request, redirect, url_for
import os
from os.path import join, dirname, realpath
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

app = Flask(__name__)

# enable debugging mode
app.config["DEBUG"] = True

# Upload folder
UPLOAD_FOLDER = 'static/files'
app.config['UPLOAD_FOLDER'] =  UPLOAD_FOLDER


# Root URL
@app.route('/')
def index():
     # Set The upload HTML template '\templates\index.html'
    return render_template('index2.html')

@app.route('/chart2.html')
def chart():
     # Set The upload HTML template '\templates\index.html'
    return render_template('chart2.html')

def parseCSV(filePath):
    TR1=0
    TR2=0
    TR3=0
    on1=0
    on2=0
    on3=0
    off1=0
    off2=0
    off3=0
      # Loop through the Rows
    D1=[]
    D2=[]
    D3=[]
    D11,D21,D31= ([] for i in range(3))
    D12,D22,D32= ([] for i in range(3))
    D13,D23,D33= ([] for i in range(3))
    D14,D24,D34= ([] for i in range(3))
    D15,D25,D35= ([] for i in range(3))

    csvData = pd.read_csv(filePath, header=None)
    
    for i,row in csvData.iterrows():
        if(i==0):
            continue
        elif(i<157):
            D11.append(i)
            D12.append(int(row[2]))
            D13.append(row[3])
            D14.append(row[4])
            D15.append(row[5])
            if(i==156):
                TR1=float(int(row[3]))
                on1=(int(row[4]))
                off1=(int(row[5]))

        elif(i<313):
            D21.append(i)
            D22.append(int(row[2]))
            D23.append(row[3])
            D24.append(row[4])
            D25.append(row[5])
            if(i==312):
                TR2=float(int(row[3]))
                on2=(int(row[4]))
                off2=(int(row[5]))

        else:
            D31.append(i)
            D32.append(int(row[2]))
            D33.append(row[3])
            D34.append(row[4])
            D35.append(row[5])
            if(i==468):
                TR3=float(int(row[3]))
                on3=(int(row[4]))
                off3=(int(row[5]))
    
    avgr1=float(TR1/float(156))
    avgr2=float(TR2/float(156))
    avgr3=float(TR3/float(156))
    
    plt.plot(D11, D12, color='green')
    plt.xlabel("Time( minutes)",fontweight ='bold', fontsize = 12)
    plt.ylabel("Rotations",fontweight ='bold', fontsize = 12)
    plt.title("Rotations vs Time device 1" ,fontweight ='bold', fontsize = 15 )
    
    plt.savefig(os.path.join('static', 'images', 'rotations1.jpg'))
    plt.clf()
    
    plt.plot(D11, D22, color='green')
    plt.xlabel("Time( minutes)",fontweight ='bold', fontsize = 12)
    plt.ylabel("Rotations",fontweight ='bold', fontsize = 12)
    plt.title("Rotations vs Time device 2" ,fontweight ='bold', fontsize = 15 )
    
    plt.savefig(os.path.join('static', 'images', 'rotations2.jpg'))
    plt.clf()
    
    plt.plot(D11, D32, color='green')
    plt.xlabel("Time( minutes)",fontweight ='bold', fontsize = 12)
    plt.ylabel("Rotations",fontweight ='bold', fontsize = 12)
    plt.title("Rotations vs Time device 3" ,fontweight ='bold', fontsize = 15 )
    
    plt.savefig(os.path.join('static', 'images', 'rotations3.jpg'))
    plt.clf()
    
    data = {'Device 1': avgr1, 'Device 2': avgr2, 'Device 3':avgr3}
    courses = list(data.keys())
    values = list(data.values())

#     fig = plt.figure(figsize = (10, 5))
#     creating the bar plot
    plt.bar(courses, values, color ='maroon',width = 0.4)
 
    plt.xlabel("DEVICES",fontweight ='bold', fontsize = 12)
    plt.ylabel("Average Rotation Per Minute",fontweight ='bold', fontsize = 12)
    plt.title("Average Rotations OF All Devices" ,fontweight ='bold', fontsize = 12 )
    plt.savefig(os.path.join('static', 'images', 'avg.jpg'))
    
    plt.clf()
    
    period1 = ['ON time', 'OFF time']
 
    data1 = [on1,off1]


    # Creating explode data
    explode = (0.0,0.1)

    # Creating color parameters
    colors = ( "red", "black")

    # Wedge properties
    wp = { 'linewidth' : 1, 'edgecolor' : "black" }

    # Creating autocpt arguments
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:.1f}%\n({:d} min)".format(pct, absolute)

    # Creating plot
    fig, ax = plt.subplots(figsize =(10, 7))
    wedges, texts, autotexts = ax.pie(data1,
                                      autopct = lambda pct: func(pct, data1),
                                      explode = explode,
                                      labels = period1,
                                      shadow = True,
                                      colors = colors,
                                      startangle = 90,
                                      wedgeprops = wp,
                                      textprops = dict(color ="white"))

    # Adding legend
    ax.legend(wedges, period1,
              loc ="center left",
              bbox_to_anchor =(1, 0, 0.5, 1))

    plt.setp(autotexts, size = 8, weight ="bold")
    plt.title("Working Period for Device 1" ,fontweight ='bold', fontsize = 15 )
    plt.savefig(os.path.join('static', 'images', 'period1.jpg'))
    plt.clf()
    
    period1 = ['ON time', 'OFF time']
 
    data1 = [on2,off2]


    # Creating explode data
    explode = (0.0,0.1)

    # Creating color parameters
    colors = ( "red", "black")

    # Wedge properties
    wp = { 'linewidth' : 1, 'edgecolor' : "black" }

    # Creating autocpt arguments
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:.1f}%\n({:d} min)".format(pct, absolute)

    # Creating plot
    fig, ax = plt.subplots(figsize =(10, 7))
    wedges, texts, autotexts = ax.pie(data1,
                                      autopct = lambda pct: func(pct, data1),
                                      explode = explode,
                                      labels = period1,
                                      shadow = True,
                                      colors = colors,
                                      startangle = 90,
                                      wedgeprops = wp,
                                      textprops = dict(color ="white"))

    # Adding legend
    ax.legend(wedges, period1,
              loc ="center left",
              bbox_to_anchor =(1, 0, 0.5, 1))

    plt.setp(autotexts, size = 8, weight ="bold")
    plt.title("Working Period for Device 2" ,fontweight ='bold', fontsize = 15 )
    plt.savefig(os.path.join('static', 'images', 'period2.jpg'))
    plt.clf()
    
    period1 = ['ON time', 'OFF time']
 
    data1 = [on3,off3]


    # Creating explode data
    explode = (0.0,0.1)

    # Creating color parameters
    colors = ( "red", "black")

    # Wedge properties
    wp = { 'linewidth' : 1, 'edgecolor' : "black" }

    # Creating autocpt arguments
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:.1f}%\n({:d} min)".format(pct, absolute)

    # Creating plot
    fig, ax = plt.subplots(figsize =(10, 7))
    wedges, texts, autotexts = ax.pie(data1,
                                      autopct = lambda pct: func(pct, data1),
                                      explode = explode,
                                      labels = period1,
                                      shadow = True,
                                      colors = colors,
                                      startangle = 90,
                                      wedgeprops = wp,
                                      textprops = dict(color ="white"))

    # Adding legend
    ax.legend(wedges, period1,
              loc ="center left",
              bbox_to_anchor =(1, 0, 0.5, 1))

    plt.setp(autotexts, size = 8, weight ="bold")
    plt.title("Working Period for Device 3" ,fontweight ='bold', fontsize = 15 )
    plt.savefig(os.path.join('static', 'images', 'period3.jpg'))
    plt.clf()
    
# Get the uploaded files
@app.route("/", methods=['POST'])
def uploadFiles():
      # get the uploaded file
    uploaded_file = request.files['file']
    if uploaded_file.filename != '':
        file_path = os.path.join(app.config['UPLOAD_FOLDER'], uploaded_file.filename)
       # set the file path
        uploaded_file.save(file_path)
        parseCSV(file_path)
       # save the file
    return redirect(url_for('chart'))


if (__name__ == "__main__"):
     app.run(port = 3010,debug=False)
```
